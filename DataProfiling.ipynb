```python
import json
import statistics
```

```python
user_names = set()
languages = set()
stars = []
open_issues = {}
min_max = {}
for readLine in open("/Users/Ramesh Srinivasan/Desktop/Spark.json", encoding="utf8"):
    dataRow = json.loads(readLine)

    print(dataRow[0])
    for element in dataRow:
        if 'username' in element:
            user_names.add(element['username'])
            
        if 'language' in element:
            language = element['language']
            languages.add(language)
            if 'open_issues' in element:
                # adding number of issues to the open_issues dictionary
                if not language in open_issues:
                    open_issues[language] = []
                
                if not language in min_max:
                    min_max[language] = {'min': element['open_issues'], 'max': element['open_issues']}
                
                if element['open_issues'] < min_max[language]['min']:
                    min_max[language]['min'] = element['open_issues']
                    
                if element['open_issues'] > min_max[language]['max']:
                    min_max[language]['max'] = element['open_issues']
                    
                open_issues[language].append(element['open_issues'])
        
        if 'stars' in element:
            stars.append(element['stars'])
    
    print('Unique user names:', len(user_names))
    print('Unique languages:', len(languages))
    print('Average stars:', statistics.mean(stars))
    print('Median stars:', statistics.median(stars))
    
    print('')
    
    for language in open_issues:
        print(f'{language} Language open issues stats')
        print('Maximum', min_max[language]['max'])
        print('Minimum', min_max[language]['min'])
        print('Average', statistics.mean(open_issues[language]))
        print('Median', statistics.median(open_issues[language]))
        print('')
```

```python
# 4. Accept subscribers and fork count from user. Display count of projects in the dataset with subscribers and fork count 
# greater than or equal to user provided input.
subscribers = int(input('Enter the number of subscribers:'))
fork_count = int(input('Enter the number of fork count:'))

count = 0

for element in dataRow:
    if 'subscribers' in element and element['subscribers'] >= subscribers and \
       'forks' in element and element['forks'] >= fork_count:
        count = count + 1

print('Count', count)
```

```python
# 5. Accept a topic of interest from user. Display count of records relevant to user's input using topics.

topic = input('Enter a topic of interest')
count = 0

for element in dataRow:
    if 'topics' in element and topic in element['topics']:
        count = count + 1
print(f'Count of projects with topic {topic}: {count}')
```
